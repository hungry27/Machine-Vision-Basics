# 5.2 Flipping an Image

```python
print("Hello World!")

img = cv2.imread("lena.png")

col_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

BGRgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(img.shape)
print(img.size)
print(img.dtype)
```

```python

plt.imshow(col_img)
plt.title("Lena Color")

```

```python
plt.imshow(BGRgray, cmap= 'gray')
plt.title ("Lena Grayscale")
```

```python

rowLength= col_img.shape[0]
colLength = col_img.shape[0]
rowLen = BGRgray.shape[0]
colLen = BGRgray.shape[0]
zeros1= np.zeros(img.shape[0:3], dtype="uint8")


for row in range(rowLength):
    zeros1[row,:,:] = img[rowLength-row-1,:,:]
    
```

```python
zeros2= np.zeros(img.shape[0:3], dtype="uint8")
for row in range(rowLength):
    for col in range(colLength):
        zeros2[:,col,:] = img[:,colLength-col-1,:]
```

```python
zeros3= np.zeros(BGRgray.shape[0:2], dtype="uint8")
for row in range(rowLength):
    for col in range(colLength):
        zeros3[:,col] = BGRgray[:,colLength-col-1]
```

```python
zeros4= np.zeros(BGRgray.shape[0:2], dtype="uint8")
for row in range(rowLength):
    zeros4[row,:] = BGRgray[rowLength-row-1,:]
```

```python
fig = plt.figure(figsize=(10,7))
rows = 2
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(cv2.cvtColor(zeros2,cv2.COLOR_BGR2RGB))
plt.axis('on')
plt.title ("Lena Color Horizontal Flip")

fig.add_subplot(rows, columns, 2)
plt.imshow(cv2.cvtColor(zeros1,cv2.COLOR_BGR2RGB))
plt.axis('on')
plt.title ("Lena Color Vertical Flip")


fig.add_subplot(rows, columns, 3)
plt.imshow(zeros3,cmap='gray')
plt.axis('on')
plt.title ("Lena Horizontal Flip")

fig.add_subplot(rows, columns, 4)
plt.imshow(zeros4,cmap='gray')
plt.axis('on')
plt.title ("Lena Vertical Flip")

```

# 5.3 Negating Image Pixels

```python
print(col_img.shape)
print(BGRgray.shape)

colneg= np.zeros(col_img.shape[0:3], dtype="uint8")
grayneg= np.zeros_like(BGRgray)

colneg.fill(511)
grayneg.fill(511)

for row in range(rowLength):
     for col in range(colLength):
            colneg[row][col] = ((colneg[row][col])- (col_img[row][col]))

```

```python
for row in range(rowLen):
     for col in range(colLen):
            grayneg[row][col] = (255 - BGRgray[row][col])
```

```python
fig3 = plt.figure(figsize=(10,7))
rows = 1
columns = 2

fig3.add_subplot(rows, columns, 1)
plt.imshow(colneg)
plt.axis('on')
plt.title ("Colored Negated Lena")

fig3.add_subplot(rows, columns, 2)
plt.imshow(grayneg, cmap = 'gray')
plt.axis('on')
plt.title ("Gray Negated Lena")
```

# 5.4 Separating Color Channels

```python
redArray = np.zeros_like(col_img)
greenArray = np.zeros_like(col_img)
blueArray = np.zeros_like(col_img)

for row in range(rowLength):
    for col in range(colLength):
        temp = col_img[row][col]
        redArray[row][col] = [temp[0],0,0]
        greenArray[row][col] = [0,temp[1],0]
        blueArray[row][col] = [0,0,temp[2]]
        
        
```

# 4 Images 1 Window


```python
fig2 = plt.figure(figsize=(10,7))
rows = 2
columns = 2

fig2.add_subplot(rows, columns, 1)
plt.imshow(redArray)
plt.axis('on')
plt.title ("Red Lena")

fig2.add_subplot(rows, columns, 2)
plt.imshow(greenArray)
plt.axis('on')
plt.title ("Green Lena")


fig2.add_subplot(rows, columns, 3)
plt.imshow(blueArray)
plt.axis('on')
plt.title ("Blue Lena")

fig2.add_subplot(rows, columns, 4)
plt.imshow(col_img)
plt.axis('on')
plt.title ("Initial Image")


```
